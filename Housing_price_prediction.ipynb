```python
import pandas as pd
```

```python
df = pd.read_csv('Housing.csv')
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['furnishingstatus'].nunique()
```

```python
columns_to_transform = ['mainroad', 'guestroom', 'basement','hotwaterheating','airconditioning','prefarea']
df[columns_to_transform] = df[columns_to_transform].replace({'yes': 1, 'no': 0})

df['furnishingstatus'] = df['furnishingstatus'].replace({'unfurnished': 0, 'semi-furnished': 1, 'furnished': 2})
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
sc = ['price', 'area']
df[sc] = scaler.fit_transform(df[sc])
```

```python
df
```

```python
df.dtypes
```

```python
core=df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(core,annot=True,cmap='Blues')
plt.show()
```

```python
df.hist(figsize=(10,10),bins=10)
plt.show()
```

```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
lr_accuracy=r2_score(y_test,y_pred)*100
```
